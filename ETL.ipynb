```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Load csvs 
csv_path_oct2020 = Path("202010-citibike-tripdata.csv")
df_oct = pd.read_csv(csv_path_oct2020, encoding="utf-8")

csv_path_nov2020 = Path("202011-citibike-tripdata.csv")
df_nov = pd.read_csv(csv_path_nov2020, encoding="utf-8")

csv_path_dec2020 = Path("202012-citibike-tripdata.csv")
df_dec = pd.read_csv(csv_path_dec2020, encoding="utf-8")
```

```python
df_oct.head()
```

```python
df_nov.head()
```

```python
df_dec.head()
```

```python
# Combine dfs
combined_df = pd.concat([df_oct, df_nov, df_dec], ignore_index=True)
combined_df
```

```python
# Update gender values 
gender_mapping = {0: "Unknown", 1: "Male", 2:"Female"}
combined_df["gender"] = combined_df["gender"].map(gender_mapping)
combined_df
```

```python
# Save to csv file 
combined_df.to_csv("combined_citibike_data.csv", encoding="utf-8", index=False, header=True)
```
